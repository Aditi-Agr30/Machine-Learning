```python
#Importing essential libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#now loading data
df = pd.read_csv('C:/Users/Aditi Agrawal/Desktop/ML_data/facebook_salary.csv')
```

```python
exp = df.iloc[:,0].values #Using only experience
```

```python
sal = df.iloc[:,-1].values #Using salary
```

```python
#Now making datasets for training and testing
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y = train_test_split(exp,sal,test_size=1/3)
```

```python
#Now calling linear regression
from sklearn.linear_model import LinearRegression
```

```python
#Calling linear regression
lr=LinearRegression()
```

```python
#Training data
lr_trained=lr.fit(train_x.reshape(-1,1),train_y.reshape(-1,1))
```

```python
#Now predicting salary
lr_trained.predict(test_x.reshape(-1,1))
```

```python
#Actual salary
test_y
```

```python
#Now visualizing output for training data
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.scatter(train_x,train_y,label="Current Salary")
plt.plot(train_x,lr_trained.predict(train_x.reshape(-1,1)),label="Predicted Salary",color="red")
plt.legend()
plt.show()
```

```python
#Now visualizing output for testing data 
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.scatter(test_x,test_y,label="Current Salary")
plt.plot(train_x,lr_trained.predict(train_x.reshape(-1,1)),label="Predicted Salary",color="red")
plt.legend()
plt.show()
```
